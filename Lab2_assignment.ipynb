# Question 1

```python
import pandas as pd
log_df = pd.read_csv("/home/datascience/wc_day6_1_sample.csv",
                     names=['ClientID', 'Date', 'Time', 'URL', 'ResponseCode', 'Size'],
                     na_values=['-'])
len(log_df[(log_df.Date == '30/Apr/1998') & (log_df.ResponseCode == 404)])
```

# Question 2

```python
may1_df = log_df[log_df['Date'] == '01/May/1998']
may1_df['DateTime'] = pd.to_datetime(may1_df.apply(lambda row: row['Date'] + ' ' + row['Time'], axis=1))
hour_grouped = may1_df.groupby(lambda x: may1_df['DateTime'][x].hour)
%matplotlib inline
ax=hour_grouped['ClientID'].nunique().plot(kind='bar')
ax.set_xlabel('Hour of the day')
ax.set_ylabel('Number of unique users')
ax.set_title('The number of distinct users of the site every hour on 01/May/1998\n')
```

# Question 3

```python
import matplotlib.pyplot as plt
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
x = hour_grouped.size().index

ax1.plot(x, hour_grouped['ClientID'].nunique(), 'g-')
ax2.plot(x, hour_grouped['Size'].sum(), 'b-')

ax1.set_xlabel('Hour of the day')
ax1.set_ylabel('# Number of unique users', color='g')
ax2.set_ylabel('Size of traffic handled', color='b')
```

##### Observations:

We can see a slight correlation between the number of unique users and size of the traffic handled.
As the number of unique users increase,the size of the traffic handled mostly increases and vice versa

# Question 4

```python
df_datetime=log_df.copy()
df_datetime=log_df[log_df['ClientID'].isin(log_df['ClientID'].unique()[0:100])]
df_datetime['DateTime'] = pd.to_datetime(df_datetime.apply(lambda row: row['Date'] + ' ' + row['Time'], axis=1))
df_datetime['Hour'] = df_datetime.apply(lambda x: x['DateTime'].hour,axis=1)
df_datetime.plot(kind='scatter', x='ClientID', y='Hour',title='The hours of the day clients sent request')
```

##### Observations:

We can see from the above graph the time at which each user sends requests. For eg user 35000 sends request at the 15th hour
We can also see at which time most clients send requests. From the above graph, it seems like a lot of  clients request data between
the 15th and 20 th hour

## 24th JULY

### A line plot of the number of distinct users vs size of traffic of the site every hour 

```python
dates_24=file_data[(file_data["Date"]=="24/Jul/1998")]
dates_24['DateTime'] = pd.to_datetime(dates_24.apply(lambda row: row['Date'] + ' ' + row['Time'], axis=1))
hour_grouped_24 = dates_24.groupby(lambda x: dates_24['DateTime'][x].hour)
%matplotlib inline
import matplotlib.pyplot as plt
fig,ax1=plt.subplots()
ax2=ax1.twinx()
x=hour_grouped_24.size().index
ax1.plot(x,hour_grouped_24['ClientID'].nunique(), 'g-')
ax2.plot(x,hour_grouped_24['Size'].sum(), 'b-')
ax1.set_xlabel('Hour of the day')
ax1.set_ylabel('# Number of unique users', color='g')
ax2.set_ylabel('Size of traffic handled', color='b')
```

##### Observations

We can see a direct correlation between the number of unique users and size of the traffic handled.
With time ,the number of unique users and the size of the traffic reduces

### Scatter plot that shows the hours of the day these clients sent request

```python
top_100=dates_24[0:100]
top_100['Hour'] = top_100.apply(lambda x: x['DateTime'].hour,axis=1)
top_100.plot(kind='scatter', x='ClientID', y='Hour')
```

##### Observations

We can see from the above graph the time at which each user sends requests. The top 100 clients selected have clients who make request within the first hour
