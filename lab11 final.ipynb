```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import math
time_len=11
degree=2

train_X=np.arange( 1,12 )
validation_X=np.array([12])
degrees = [1, 2, 3,4]
from sklearn import cross_validation
for i in range(len(degrees)):
    sum=0
    for k in et:
        cols=[]
        colY=[]
        colY.append('%d%s'%(2014,k))
        for j in range(2003, 2014):
            cols.append('%d%s'%(j,k))

        train_Y= np.array(df_pt_final[cols])
        train_Y = train_Y.reshape(time_len, -1)
        validation_Y=np.array(df_pt_final[colY])
        polynomial_features = PolynomialFeatures(degree=degrees[i],
                                                 include_bias=False)
        linear_regression = LinearRegression()
        pipeline = Pipeline([("polynomial_features", polynomial_features),
                             ("linear_regression", linear_regression)])
        pipeline.fit(train_X.reshape(11,-1), train_Y)
        
        predict=pipeline.predict(validation_X)
        E=(predict.transpose()-validation_Y) ** 2
        sum=sum+np.sum(E)
        
    print("Degree :",degrees[i] ,"Avg Error", math.sqrt(sum/(len(E)*6)))
```
